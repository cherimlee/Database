```python
from konlpy.tag import Okt, Komoran, Kkma, Hannanum, Twitter
okt = Okt()
tokens = okt.morphs(text)
tokens
```

```python
wholeemo = pd.read_excel('./정연_채림_emotion (1).xlsx')
wholeemo = wholeemo.rename(columns={'단어':'morphs'})
wholeemo.set_index('morphs', inplace=True)
wholeemo
```

```python
sent_words = []

for tok in tokens:
    try:
        ser = wholeemo.loc[tok]
#         print(ser[wholeemo.loc[tok] == 1].index[0], end=' ')
        sent_words.append(ser[wholeemo.loc[tok] == 1].index[0])

    except:
        pass
pd.Series(sent_words).value_counts()
```
